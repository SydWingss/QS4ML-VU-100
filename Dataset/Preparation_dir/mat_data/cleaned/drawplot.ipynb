```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import os
from sklearn.preprocessing import MaxAbsScaler
```

```python
df= pd.read_csv('sensorlog_20240606_121611 synthesis.csv')
columns_to_exclude = ['altitude', 'course', 'hacc', 'latitude', 'longitude', 'speed']

# 生成一个新的DataFrame，只包含要绘制的列
df= df.drop(columns=columns_to_exclude)
df.rename(columns={'Timestamp': 'dateTime'}, inplace=True)
df2 = pd.read_csv('C:\\Users\\sydwi\\Desktop\\QSML\\data\\QS4ML-VU-100\\band_data\\cleaned\\band-syn.csv')
df= pd.merge(df, df2, how='outer')
df=df.drop(columns='dateTime')
max_abs_scaler = MaxAbsScaler()
df.rename(columns={'rate': 'heart_rate'}, inplace=True)
df.rename(columns={'BandAccX': 'BandAcceleration_X'}, inplace=True)
df.rename(columns={'BandAccY': 'BandAcceleration_Y'}, inplace=True)
df.rename(columns={'BandAccZ': 'BandAcceleration_Z'}, inplace=True)

# 对DataFrame进行最大绝对值缩放
df = pd.DataFrame(max_abs_scaler.fit_transform(df), columns=df.columns)

```

```python
sample_df = df.sample(frac=0.1, random_state=1)
plt.figure(figsize=(10, 6))
sample_df.boxplot()
plt.title('Box Plot of Motion Sensors')
plt.ylabel('Values')
plt.xticks(rotation=45, ha='right')

# 调整标签大小
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.tight_layout()  # 自动调整子图参数，以适应标签




plt.show()
```

```python
# 抽样
sample_df = df.sample(frac=0.1, random_state=1)

# 使用seaborn样式
sns.set(style="whitegrid")

# 绘制箱形图
plt.figure(figsize=(12, 8))
boxplot = sample_df.boxplot()
plt.title('Box Plot of Motion Sensors', fontsize=16, fontweight='bold')
plt.ylabel('Values', fontsize=14)
plt.xlabel('Sensors', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.grid(True, linestyle='--', linewidth=0.7)

# 自动调整子图参数，以适应标签
plt.tight_layout()

# 显示图表
plt.show()
```

```python
sample_df = df.sample(frac=0.1, random_state=1)

# Use seaborn style
sns.set(style="whitegrid")

# Create a box plot with colors
plt.figure(figsize=(12, 8))
boxplot = sample_df.boxplot(patch_artist=True,
                            boxprops=dict(facecolor='lightblue', color='blue'),
                            whiskerprops=dict(color='blue'),
                            capprops=dict(color='blue'),
                            medianprops=dict(color='red'),
                            flierprops=dict(markerfacecolor='red', marker='o', markersize=5, linestyle='none'))

# Add title and labels
plt.title('Box Plot of Motion Sensors', fontsize=16, fontweight='bold')
plt.ylabel('Values', fontsize=14)
plt.xlabel('Sensors', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.grid(True, linestyle='--', linewidth=0.7)

# Adjust layout to fit labels
plt.tight_layout()

# Show plot
plt.show()
```
