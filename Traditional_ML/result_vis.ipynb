```python
import matplotlib.pyplot as plt
import numpy as np

# 假设你有模型名称和性能的数据
models = ['RF', 'RF_RFE', 'XGB_RFE']
accuracy = [0.182, 0.258, 0.455]
f_1 = [0.041, 0.197, 0.344]

# 定义颜色
colors_accuracy = ['#1f77b4', '#2e86c1', '#2874a6']
colors_f1 = ['#aec7e8', '#b0c4de', '#99badd']

# 创建图表
plt.style.use('seaborn-v0_8-white')
fig, ax = plt.subplots(figsize=(12, 6))  # 在这里设置图形大小

# 设置条形宽度
bar_width = 0.2

# 计算每个柱的位置
r1 = np.arange(len(models))
r2 = [x + bar_width for x in r1]

# 绘制条形
bars1 = ax.bar(r1, accuracy, color=colors_accuracy, width=bar_width, label='Accuracy')
bars2 = ax.bar(r2, f_1, color=colors_f1, width=bar_width, label='F1 Score')

# 在每个条形上方显示具体数值
for bar in bars1:
    height = bar.get_height()
    ax.annotate(f'{height:.3f}',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3),  # 3 points vertical offset
                textcoords="offset points",
                ha='center', va='bottom')

for bar in bars2:
    height = bar.get_height()
    ax.annotate(f'{height:.3f}',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3),  # 3 points vertical offset
                textcoords="offset points",
                ha='center', va='bottom')

# 添加标签和标题
ax.set_xlabel('Models', fontweight='bold')
ax.set_ylabel('Scores', fontweight='bold')
ax.set_title('Model Performance', fontweight='bold')
ax.set_xticks([r + bar_width/2 for r in range(len(models))])
ax.set_xticklabels(models)

# 去除纵向网格线
ax.yaxis.grid(True)
ax.xaxis.grid(False)
ax.set_ylim(0, 0.6)

# 保留一定的页边距
plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1)

# 添加图例
ax.legend()

# 显示图表
plt.show()
```
